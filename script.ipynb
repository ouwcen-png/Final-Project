```python
import pandas as pd

# Load the dataset
def load_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Clean the dataset
def clean_data(data):
    # Example cleaning: Remove missing values and drop irrelevant columns
    data_cleaned = data.dropna()  # Remove rows with missing values
    data_cleaned = data_cleaned[['Year', 'Location', 'Performer', 'Role']]  # Keep relevant columns
    return data_cleaned

# Save cleaned data
def save_cleaned_data(data, output_path):
    data.to_csv(output_path, index=False)

if __name__ == "__main__":
    file_path = 'data/kunqu_opera_data.csv'  # Example raw data path
    output_path = 'data/cleaned_kunqu_opera_data.csv'

    # Load, clean, and save the dataset
    data = load_data(file_path)
    cleaned_data = clean_data(data)
    save_cleaned_data(cleaned_data, output_path)
    print(f"Cleaned data saved to {output_path}")
```

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize Sentiment Intensity Analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Load the cleaned dataset
def load_cleaned_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Perform sentiment analysis on the 'Role' column (you can modify this to other text columns)
def analyze_sentiment(data):
    data['Sentiment'] = data['Role'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
    return data

# Save the sentiment results
def save_sentiment_data(data, output_path):
    data.to_csv(output_path, index=False)

if __name__ == "__main__":
    file_path = 'data/cleaned_kunqu_opera_data.csv'  # Path to the cleaned data
    output_path = 'data/sentiment_kunqu_opera_data.csv'

    # Load the data, perform sentiment analysis, and save the results
    data = load_cleaned_data(file_path)
    sentiment_data = analyze_sentiment(data)
    save_sentiment_data(sentiment_data, output_path)
    print(f"Sentiment analysis results saved to {output_path}")
```

```python
import pandas as pd
import plotly.express as px

# Load the sentiment data
def load_sentiment_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Create sentiment distribution plot
def plot_sentiment_distribution(data):
    fig = px.histogram(data, x="Sentiment", title="Sentiment Distribution of Kunqu Opera Roles")
    fig.update_layout(bargap=0.1)
    fig.show()

# Create role participation plot
def plot_role_participation(data):
    role_counts = data['Role'].value_counts()
    fig = px.bar(role_counts, x=role_counts.index, y=role_counts.values, title="Role Participation in Kunqu Opera")
    fig.update_layout(xaxis_title="Role", yaxis_title="Frequency")
    fig.show()

if __name__ == "__main__":
    file_path = 'data/sentiment_kunqu_opera_data.csv'  # Path to the sentiment analyzed data

    # Load the sentiment data
    data = load_sentiment_data(file_path)

    # Plot the sentiment distribution and role participation
    plot_sentiment_distribution(data)
    plot_role_participation(data)
```
